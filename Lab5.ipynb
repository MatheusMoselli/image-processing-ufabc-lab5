*Elabore um programa para realizar continuamente sobre a imagem da WEBCAM as operações de correção de ponto branco, e a correção gama, mostrando na tela o resultado em tempo real, permitindo ao operador escolher a visualização da imagem original, correção ponto branco, ou correção gama, ou ambas. Grave um video demonstrativo com os integrantes do grupo.*

```python
import cv2
import numpy as np

def white_point_correction(img, sourcewp, targetwp):
    # img escala [0, 1]
    corrected = img * (targetwp / sourcewp)
    return np.clip(corrected, 0, 1)

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Não foi possível abrir a câmera")
    exit()

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    # if frame is read correctly ret is True
    if not ret:
        print("Não foi possível receber o frame (stream end?). Finalizando...")
        break

    # Corrigindo de uma luz quente (amarelada) para D65 (luz do dia)
    img_rgb = frame / 255.0
    
    # Simplificação: canais R, G, B
    source_wp = np.array([0.8, 0.8, 0.5])
    target_wp = np.array([1.0, 1.0, 1.0])
    corrected_img = white_point_correction(img_rgb, source_wp, target_wp)
    
    # Display the resulting frame
    cv2.imshow('frame', corrected_img)
    
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()        
```

*Elabore um programa para realizar continuamente sobre a imagem da WEBCAM a detecção de uma cor
pelo método HSV, mostrando lado-a-lado na tela a imagem original, e a imagem somente com as partes
detectadas. Grave um video demonstrativo com os integrantes do grupo.*

```python
import cv2
import numpy as np

def segmentar_por_cor(img):
    # 1. Converter para HSV
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # 2. Definir o intervalo da cor desejada (Exemplo: Verde)
    # No OpenCV: H vai de 0 a 179, S e V de 0 a 255
    lower_blue = np.array([100, 50, 50])
    upper_blue = np.array([130, 255, 255])


    # 3. Criar uma máscara binária
    mask = cv2.inRange(hsv, lower_blue, upper_blue)

    # 4. Aplicar a máscara na imagem original
    resultado = cv2.bitwise_and(img, img, mask=mask)

    return mask, resultado

cap = cv2.VideoCapture(0)

width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

fps = 60.0 # alterar pra um valor alto pra capturar os videos rápidos

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out_original = cv2.VideoWriter('videos_coletados/saida_original.avi', fourcc, fps, (int(width), int(height)))
out_resultado = cv2.VideoWriter('videos_coletados/saida_resultado.avi', fourcc, fps, (int(width), int(height)))


if not cap.isOpened():
    print("Não foi possível abrir a câmera")
    exit()
    
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    # if frame is read correctly ret is True
    if not ret:
        print("Não foi possível receber o frame (stream end?). Finalizando...")
        break

    mascara, resultado = segmentar_por_cor(frame)
    
    # Display the resulting frame
    out_original.write(frame)
    out_resultado.write(resultado)
    cv2.imshow('Original', frame)
    cv2.imshow('Resultado', resultado)
    
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
out_original.release()
out_resultado.release()
cv2.destroyAllWindows()
```
