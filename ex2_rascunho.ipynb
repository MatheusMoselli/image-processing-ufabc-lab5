```python
#EX2 - rascunho

import cv2 as cv
import numpy as np

IMG_NAMES = ["messi5.jpg", "images.jpeg"]

def read_or_warn(path: Path):
#Ajustar mecanismo de path depois com as funções genericas do notebook
    img = cv.imread(str(path))
    if img is None:
        print(f"Não foi possível abrir: {path}")
    return img


#Quebrado, arrumar
def aplicar_gamma(image, gamma):
    inv_gamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** inv_gamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
   
    return cv.LUT(image, table.reshape((256, 1)))

for name in IMG_NAMES:
    img = read_or_warn(name)
    img_corrigida = aplicar_gamma(img, 2.2)

    cv.imshow('Imagem com correção inversa aplicada', img_corrigida)
    cv.waitKey(0)
    cv.destroyAllWindows()
```
