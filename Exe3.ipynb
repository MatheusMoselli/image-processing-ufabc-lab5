```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def color_to_grayscale_ycbcr(img_path):
    img_bgr = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # Converter para YCbCr
    img_ycbcr = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2YCrCb).astype(np.float32)

    # Zerar canais de crominância (Cb e Cr = 128)
    img_ycbcr[:, :, 1] = 128  # Cr
    img_ycbcr[:, :, 2] = 128  # Cb

    # Converter de volta para RGB
    img_ycbcr = np.clip(img_ycbcr, 0, 255).astype(np.uint8)
    img_gray_rgb = cv2.cvtColor(img_ycbcr, cv2.COLOR_YCrCb2RGB)

    # Visualização
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    axes[0].imshow(img_rgb)
    axes[0].set_title('Imagem Original')
    axes[0].axis('off')

    axes[1].imshow(img_gray_rgb)
    axes[1].set_title('Imagem em Escala de Cinza\n(Cb = Cr = 128)')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# Execução
color_to_grayscale_ycbcr('img/alunos.png')
```
