```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def apply_inverse_gamma(img_path, gamma=2.2):
    img_bgr = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_normalized = img_rgb.astype(np.float32) / 255.0

    # Aplicar correção gama inversa
    img_gamma_inverse = np.power(img_normalized, gamma)
    img_gamma_inverse = np.clip(img_gamma_inverse, 0, 1)

    # Visualização
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    axes[0].imshow(img_normalized)
    axes[0].set_title('Imagem Original')
    axes[0].axis('off')

    axes[1].imshow(img_gamma_inverse)
    axes[1].set_title(f'Correção Gama Inversa (γ = {gamma})')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# Execução
apply_inverse_gamma('img/avatares.png', gamma=2.2)
```
