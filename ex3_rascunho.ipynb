```python
#EX3

import cv2 as cv
import numpy as np

IMG_NAMES = ["messi5.jpg"]

def read_or_warn(path: Path):
#Ajustar mecanismo de path depois com as funções genericas do notebook
    img = cv.imread(str(path))
    if img is None:
        print(f"Não foi possível abrir: {path}")
    return img

for name in IMG_NAMES:
  img = read_or_warn(name)
  img_ycbcr = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

  #Zerando os canais
  img_ycbcr[:, :, 1] = 128  # Cr neutro
  img_ycbcr[:, :, 2] = 128  # Cb neutro

  #Reconstruida pra RGB
  img_gray_reconstructed = cv2.cvtColor(img_ycbcr, cv2.COLOR_YCrCb2BGR)

  cv2.imshow('Escala de Cinza via YCbCr', img_gray_reconstructed)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
```
